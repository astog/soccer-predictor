```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("database.sqlite")
df = pd.read_sql_query("select * from Match;", conn)
conn.close()
```

```python
odds_col = [col for col in df.columns if col.isupper()]
print(odds_col)
```

```python
df.columns.values
```

```python
keep_cols = ['home_team_goal', 'away_team_goal'] + odds_col
drop_cols = [col for col in df.columns if col not in keep_cols]
print(drop_cols)
```

```python
df.drop(drop_cols, axis=1, inplace=True)
df.head()
```

```python
print(df.shape)
df.isna().sum()
```

```python
df.dropna(thresh=21, inplace=True)
print(df.shape)
df.isna().sum()
```

```python
df.loc[:, odds_col] = 1.0 / df.loc[:, odds_col]
df.tail()
```

```python
home_cols = [col for col in df.columns if col[-1] == 'H']
home_cols
```

```python
df.loc[:, home_cols].tail()
```

```python
home_means = df.loc[:, home_cols].mean(axis=1)
home_means.tail()
```

```python
for col in home_cols:
    df.loc[:, col].fillna(df.loc[:, home_cols].mean(axis=1), inplace=True)
    
df[home_cols].tail()
```

```python
draw_cols = [col for col in df.columns if col[-1] == 'D']
draw_cols
```

```python
df.loc[:, draw_cols].tail()
```

```python
for col in draw_cols:
    df.loc[:, col].fillna(df.loc[:, draw_cols].mean(axis=1), inplace=True)
    
df[draw_cols].tail()
```

```python
away_cols = [col for col in df.columns if col[-1] == 'A']
away_cols
```

```python
df.loc[:, away_cols].tail()
```

```python
for col in away_cols:
    df.loc[:, col].fillna(df.loc[:, away_cols].mean(axis=1), inplace=True)
    
df[away_cols].tail()
```

```python
df.isna().sum()
```

```python
def get_match_result(row):
    if row['home_team_goal'] == row['away_team_goal']:
        return 0
    elif row['home_team_goal'] > row['away_team_goal']:
        return 1
    else:
        return 2
```

```python
i = 102
print(df.iloc[i])
print(get_match_result(df.iloc[i]))
```

```python
df['match_result'] = df.apply(get_match_result, axis=1)
```

```python
df.isna().sum()
```

```python
df.drop(['home_team_goal', 'away_team_goal'], axis=1, inplace=True)
```
